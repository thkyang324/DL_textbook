# Activation Function

```python
x = np.arange(-10,10,0.1)
```

```python
def step_function(x):
    output = x>0
    return output.astype(np.int)
def sigmoid(x):
    return 1/(1+np.exp(-x))
def relu(x):
    return np.maximum(0,x)
def softmax(x):
    exps = np.exp(x)
    return np.exp(x)/np.sum(exps)
```

```python
plt.plot(x,step_function(x))
plt.plot(x,sigmoid(x))
plt.plot(x,relu(x))
plt.plot(x,softmax(x))
```
